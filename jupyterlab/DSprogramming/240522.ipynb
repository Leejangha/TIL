```python
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import yfinance as yf
start_date = "2020-01-01"
end_date = "2024-01-01"
samsung = yf.download("005930.KS", start=start_date, end=end_date)
lg = yf.download("066570.KS", start=start_date, end=end_date)
tcs = yf.download("TCS.NS", start=start_date, end=end_date)
```

```python
print(samsung.head())
print(lg.head())
print(tcs.head())
```

```python
#The Graph for the Volume Traded (거래량그래프)
samsung['Volume'].plot(label='Samsung', figsize=(10, 5))
lg['Volume'].plot(label='LG')
tcs['Volume'].plot(label='TCS')
plt.title('Volume Traded')
plt.legend()
plt.show()
```

```python
#The Graph for the Close Price (종가그래프)
samsung['Close'].plot(label='Samsung', figsize=(10, 5))
lg['Close'].plot(label='LG')
tcs['Close'].plot(label='TCS')
plt.title('Close Price')
plt.legend()
plt.show()
```

```python
#The Graph for the Market Capital (시가총액그래프)
samsung['Market Cap'] = samsung['Close'] * samsung['Volume']
lg['Market Cap'] = lg['Close'] * lg['Volume']
tcs['Market Cap'] = tcs['Close'] * tcs['Volume']
samsung['Market Cap'].plot(label='Samsung', figsize=(10, 5))
lg['Market Cap'].plot(label='LG')
tcs['Market Cap'].plot(label='TCS')
plt.title('Market Cap')
plt.legend()
plt.show()
```
