```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial
```

```python
lang = "PYTHON"
xs = list(lang)
ys = list(map(ord, xs))

plt.bar(xs, ys, color='blue', width=0.7)
plt.xlabel("Charcters")
plt.ylabel("Weight")
plt.title("PYTHON Bar Chart")
```

```python
xs = list(range(10))
y1 = list(map(factorial, xs))
y2 = list(map(lambda x: 4 ** x, xs))
plt.plot(xs, y1)
plt.plot(xs, y2, '-.')
plt.xlabel("Input Size")
plt.ylabel("Computation Time")
plt.title("Power Four vs Factorial")
```
