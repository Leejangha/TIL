```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
fifa_filepath = "data/fifa.csv"

# Read the file into a variable fifa_data
fifa_data = pd.read_csv(fifa_filepath, index_col="Date", parse_dates=True)
```

```python
# Print the first 5 rows of the data
fifa_data.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Line chart showing how FIFA rankings evolved over time 
sns.lineplot(data=fifa_data)
```

```python
flight_data = pd.read_csv("data/flight_delays.csv", index_col="Month")
```

```python
flight_data.head()
```

```python
flight_data
```

```python
plt.figure(figsize=(10,6))
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")
sns.barplot(x=flight_data.index, y=flight_data['NK'])
plt.ylabel("Arrival delay (in minutes)")
```

```python
plt.figure(figsize=(14,7))
plt.title("Average Arrival Delay for Each Airline, by Month")
sns.heatmap(data=flight_data, annot=True)
plt.xlabel("Airline")
```

```python
ign_data = pd.read_csv("data/ign_scores.csv", index_col="Platform")
```

```python
ign_data
```

```python
plt.figure(figsize=(10,6))
plt.title("Average score for racing games, for each platform")
sns.barplot(x=ign_data['Racing'], y=ign_data.index)
plt.xlabel("")
```
