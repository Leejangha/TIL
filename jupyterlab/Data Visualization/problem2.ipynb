```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
austin_weather_data = pd.read_csv("data/austin_weather.csv")
austin_weather_data.head()
```

```python
df = austin_weather_data
df["Date"] = pd.to_datetime(df["Date"])

plt.figure(figsize=(12,7))
plt.title('Temperature Variation')
plt.plot(df['Date'], df['TempHighF'], label='HighTemperature')
plt.plot(df['Date'], df['TempAvgF'], label='AvgTemperature')
plt.plot(df['Date'], df['TempLowF'], label='LowTemperature')
plt.xlabel('Date')
plt.ylabel('Temperature (Fahrenehit)')
plt.legend(loc = 8)
plt.show()
```
